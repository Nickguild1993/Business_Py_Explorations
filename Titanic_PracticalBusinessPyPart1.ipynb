#### Custom Functions for aggregations

Calculate the 25th percentile of the data using multiple approaches

```python
# Using partial function

from functools import partial

# use partial 

q_25 = partial(pd.Series.quantile, q= 0.25)
q_25.__name__ = "25%"
```

```python
# define own function 

def percentile_25(x):
  return x.quantile(.25)
```

```python
# define a lambda function

lambda_25 = lambda x : x.quantile(.25)
lambda_25.__name__ = 'lambda_25%'
```

```python
# using a lambda function inline w/ the others

agg_func = {
    "Fare" : [q_25, percentile_25, lambda_25, lambda x : x.quantile(.25)]  
}

df.groupby(["Embarked"]).agg(agg_func)
```

```python
# testing it out 

def quartile_50(x):
  return x.quantile(.50)

agg_func = {
    "Fare" : [quartile_50]
}

df.groupby("Pclass").agg(agg_func).round(2)
```

#### More custom functions

#### Custom functions with multiple columns

When you want to run multiple aggregations across columns, use groupyby() combined with .apply() 

```python
def summary(x):
  result = {
      "Fair_sum" : x["Fare"].sum(),
      "Fare_mean" : x["Fare"].mean(),
      "Fare_range" : x["Fare"].max() - x["Fare"].min()
  }
  return pd.Series(result).round(0)

df.groupby(["Pclass"]).apply(summary)
```

#### Working with group objects - crosstab and pivot tables too!

Once you group and aggregate the data, you can do additional calculations on the grouped objects

Example: figure out what percentage of the total fares sold can be attributed to each town that passengers go on the boat from.

Using **assign** and a **lambda** function to add pct_total column

```python
df.groupby(["Embarked", "Pclass"]).agg({
    "Fare" : "sum"
}).assign(pct_total = lambda x: x / x.sum()).round(3)
```

That said- can do the above **MUCH** easier using a pd.crosstab

```python
pd.crosstab(df["Embarked"],
            df["Pclass"],
            values = df["Fare"],
            aggfunc = "sum",
            normalize = True)
```

Can also use agg functions with pivot tables!!!

```python
def top_10_sum(x):
    return x.nlargest(10).sum()

def bottom_10_sum(x):
    return x.nsmallest(10).sum()


agg_func_top_bottom_sum = {
    'Fare': [top_10_sum, bottom_10_sum]
}

pd.pivot_table(data = df,
               index = ["Embarked"],
               columns = ["Pclass"],
               aggfunc = agg_func_top_bottom_sum)
```

### When you have to use multiple groupby() 

Imagine, if you will.  We wanted to get a cumulative total of the fares - can group and aggregate by town and by class then group the resulting object and calculate the cumulative sum

```python
fare_group = df.groupby(["Embarked", "Pclass"]).agg({"Fare": "sum"})
fare_group.groupby(level=0).cumsum()
```

```python
# sort of tricky, but this is whats going on

df.groupby(["Embarked"]).agg({"Fare":"sum"}) # then you're just adding the .cumsum() argument with an additional groupby element -> Pclass
```
